```python
train_df, valid_df = train_test_split(df, test_size = 0.20, random_state = 42)
```

```python
train_augs = A.Compose([
    A.Resize(width=120, height=120),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
    A.Rotate()
], bbox_params=A.BboxParams(format='pascal_voc'))

valid_augs = A.Compose([
    A.Resize(width=120, height=120),
], bbox_params=A.BboxParams(format='pascal_voc'))
```

```python
class ObjLocalizationDataset(Dataset):
    
    def __init__(self, df, augmentations = None):
        self.df = df
        self.augmentations = augmentations
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        
        row = self.df.iloc[idx]
        
        xmin = row.xmin
        ymin = row.ymin
        xmax = row.xmax
        ymax = row.ymax
        label = row.label
        
        bbox = [[xmin, ymin, xmax, ymax, label]]
        
        img_path = row.img_path 
        img = cv2.imread(img_path,cv2.COLOR_BGR2RGB)
        
        if self.augmentations:
            data = self.augmentations(image = img, bboxes=bbox)
            img = data['image']
            bbox = data['bboxes']
            bbox = bbox[0]
            

        img = torch.from_numpy(img).permute(2, 0, 1) / 255.0
        label = bbox[4]
        bbox = torch.Tensor(bbox[0:4])
        
        return img, bbox, label
```

```python
trainset = ObjLocalizationDataset(train_df, train_augs)
validset = ObjLocalizationDataset(valid_df, valid_augs)
```

```python
img, bbox, label = trainset[5]

xmin, ymin, xmax, ymax = bbox

pt1 = (int(xmin), int(ymin))
pt2 = (int(xmax), int(ymax))

bnd_img = cv2.rectangle(img.permute(1, 2, 0).numpy(),pt1, pt2,(255,0,0),2)
plt.imshow(img.permute(1, 2, 0))
```

```python
trainloader = DataLoader(trainset, batch_size = 64, shuffle = True)
validloader = DataLoader(validset, batch_size = 64)
```

```python
class ObjLocModel(nn.Module):

    def __init__(self):
        super(ObjLocModel, self).__init__()
        
        self.feature_extractor = timm.create_model('resnet34', pretrained = True)
        self.feature_extractor.fc = nn.Linear(512, 256)
        
        self.classifier = nn.Sequential(
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 3)
        )
        
        self.localizer = nn.Sequential(
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 4)
        )
        
    def forward(self, images, gt_bboxes = None, labels = None):
        x = self.backbone(images)
        logits = self.classifier(x)
        bboxes = self.localizer(x)
        
        if labels != None:
            loss = nn.CrossEntropyLoss()(logits, labels) + nn.MSELoss()(bboxes, gt_bboxes)
            return logits, bboxes, loss
```
