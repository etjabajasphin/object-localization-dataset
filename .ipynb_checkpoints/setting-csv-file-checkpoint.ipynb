```python
import numpy as np 
import pandas as pd 
from xml.dom import minidom
import glob
```

```python
def get_value(obj,name) :
    a = int(obj.getElementsByTagName(name)[0].firstChild.nodeValue)
    return a
```

```python
okay_list = glob.glob('train_images/**.xml')
okay_list_2 = glob.glob('train_images/**.jpg')
```

```python
okay_list[6][:-4]
```

```python
temp_dict = {}

temp_path = []
temp_xmin = []
temp_xmax = []
temp_ymin = []
temp_ymax = []
temp_width = []
temp_height = []
temp_name = []

for i in okay_list:
    path = i[:-4] + '.jpg'
    obj = minidom.parse(i)
    xmin = get_value(obj,'xmin')
    ymin = get_value(obj,'ymin')
    xmax = get_value(obj,'xmax')
    ymax = get_value(obj,'ymax')
    width = get_value(obj,'width')
    height = get_value(obj,'height')
    
    temp_path.append(path)
    temp_xmin.append(xmin)
    temp_xmax.append(xmax)
    temp_ymin.append(ymin)
    temp_ymax.append(ymax)
    temp_width.append(width)
    temp_height.append(height)
    if "cucumber" in path:
        temp_name.append("cucumber")
    elif "eggplant" in path:
        temp_name.append("eggplant")
    elif "mushroom" in path:
        temp_name.append("mushroom")
        
    
temp_dict['img_path'] = temp_path
temp_dict['xmin'] = temp_xmin
temp_dict['ymin'] = temp_ymin
temp_dict['xmax'] = temp_xmax
temp_dict['ymax'] = temp_ymax
temp_dict['width'] = temp_width
temp_dict['height'] = temp_height
temp_dict['label'] = temp_name
```

```python
data = pd.DataFrame(temp_dict)
```

```python
import cv2
import matplotlib.pyplot as plt 
```

```python
row = data.iloc[1]
img = cv2.imread(row.img_path,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
bnd_img = cv2.rectangle(img,(row.xmin,row.ymin),(row.xmax,row.ymax),(255,0,0),2)
plt.imshow(bnd_img)
```
